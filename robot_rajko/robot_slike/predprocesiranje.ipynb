```python
import os
import cv2

# Funkcija za isecanje donje polovine slike
def crop_bottom_half(image_path):
    # Učitavanje slike
    image = cv2.imread(image_path)
    
    # Dobijanje dimenzija slike
    height, width, _ = image.shape
    
    # Određivanje sredine po visini
    mid_height = height // 2
    
    # Isecanje donje polovine slike
    cropped_image = image[mid_height:height, 0:width]
    
    return cropped_image

# Funkcija za obradu svih slika u zadatom folderu
def process_images_in_folder(folder_path):
    for filename in os.listdir(folder_path):
        # Provjera da li je fajl slika
        if filename.endswith(".png") or filename.endswith(".jpg") or filename.endswith(".jpeg"):
            image_path = os.path.join(folder_path, filename)
            cropped_image = crop_bottom_half(image_path)
            
            # Definisanje novog imena za isečenu sliku
            new_filename = "cropped_" + filename
            new_image_path = os.path.join(folder_path, new_filename)
            
            # Čuvanje isečene slike
            cv2.imwrite(new_image_path, cropped_image)

# Definisanje putanja ka folderima
ok_front_folder = 'data_set/train/ok_front'
def_front_folder = 'data_set/train/def_front'

# Obrada slika u oba foldera
process_images_in_folder(ok_front_folder)
process_images_in_folder(def_front_folder)
```

```python

# Funkcija za isecanje slike sa prilagodljivim procentima za levo i desno
def crop_left_right(image_path, left_percent, right_percent):
    # Učitavanje slike
    image = cv2.imread(image_path)
    
    # Dobijanje dimenzija slike
    height, width, _ = image.shape
    
    # Izračunavanje procenta za sečenje
    left_width = int(width * left_percent)
    right_width = int(width * right_percent)
    
    # Isecanje levo i desno
    cropped_image = image[:, left_width:width-right_width]
    
    return cropped_image

# Funkcija za obradu svih slika u zadatom folderu
def process_images_in_folder(folder_path, left_percent=0.35, right_percent=0.35):
    for filename in os.listdir(folder_path):
        # Provjera da li je fajl slika
        if filename.endswith(".png") or filename.endswith(".jpg") or filename.endswith(".jpeg"):
            image_path = os.path.join(folder_path, filename)
            cropped_image = crop_left_right(image_path, left_percent, right_percent)
            
            # Definisanje novog imena za isečenu sliku
            new_filename = "cropped_" + filename
            new_image_path = os.path.join(folder_path, new_filename)
            
            # Čuvanje isečene slike
            cv2.imwrite(new_image_path, cropped_image)

# Definisanje putanja ka folderima
ok_front_folder = 'data_set/train/ok_front'
def_front_folder = 'data_set/train/def_front'

# Obrada slika u 'ok_front' folderu (35% sa leve i desne strane)
process_images_in_folder(ok_front_folder, left_percent=0.35, right_percent=0.35)

# Obrada slika u 'def_front' folderu (25% sa leve i 45% sa desne strane)
process_images_in_folder(def_front_folder, left_percent=0.25, right_percent=0.45)
```
