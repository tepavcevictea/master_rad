```python
import os

import re

# Define the directory where your images are located
directory = 'generated_images3'

# Get a list of all files in the directory
files = os.listdir(directory)

# Define a pattern to match the old filenames
pattern = re.compile(r'defect_image_(\d+)')

# Initialize a starting number for the new filenames
start_number = 9998

# Iterate over the files in the directory
for filename in files:
    # Match the pattern to find old filenames
    match = pattern.match(filename)
    if match:
        # Extract the old image number
        old_number = int(match.group(1))
        
        # Calculate the new number based on the starting number
        new_number = start_number + old_number
        
        # Define the new filename with the desired format and extension
        new_filename = f'cast_def_0_{new_number:05d}.png'  # Change extension to '.png' if needed
        
        # Get the full path for the old and new filenames
        old_file = os.path.join(directory, filename)
        new_file = os.path.join(directory, new_filename)
        
        # Rename the file
        os.rename(old_file, new_file)
        print(f'Renamed: {old_file} -> {new_file}')

print('Renaming completed.')
```
