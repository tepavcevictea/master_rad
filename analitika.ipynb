## Analiza baze podataka

```python
import os
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

### Putanje do direktorijuma sa slikama

```python
casting_data_path = 'data_set/casting_data/casting_data'
casting_data_train_defective_path = os.path.join(casting_data_path, 'train/def_front')
casting_data_train_ok_path = os.path.join(casting_data_path, 'train/ok_front')
casting_data_test_defective_path = os.path.join(casting_data_path, 'test/def_front')
casting_data_test_ok_path = os.path.join(casting_data_path, 'test/ok_front')
```

```python
casting_512x512_path = 'data_set/casting_512x512/casting_512x512'
casting_512x512_defective_path = os.path.join(casting_512x512_path, 'def_front')
casting_512x512_ok_path = os.path.join(casting_512x512_path, 'ok_front')
```

### Ispisivaje broja slika u direktorijumu


```python
def get_image_count(directory):
    return len(os.listdir(directory))

# Ispis broja slika za svaki direktorijum
counts = {
    'Train Defective': get_image_count(casting_data_train_defective_path),
    'Train OK': get_image_count(casting_data_train_ok_path),
    'Test Defective': get_image_count(casting_data_test_defective_path),
    'Test OK': get_image_count(casting_data_test_ok_path),
    '512x512 Defective': get_image_count(casting_512x512_defective_path),
    '512x512 OK': get_image_count(casting_512x512_ok_path)
}

# Ispis rezultata
for label, count in counts.items():
    print(f"{label}: {count} slika")
```

### Generisanje grafikona sa brojem slika u svakom direktorijumu

```python
# Generisanje grafikona sa brojem slika u svakom direktorijumu
def plot_image_counts(counts):
    directories = list(counts.keys())
    image_counts = list(counts.values())
    
    plt.figure(figsize=(12, 6))
    plt.bar(directories, image_counts, color=['blue', 'orange', 'green', 'red', 'purple', 'brown'])
    plt.xlabel('Direktorijum')
    plt.ylabel('Broj slika')
    plt.title('Broj slika u svakom direktorijumu')
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y')
    plt.show()

plot_image_counts(counts)
```
